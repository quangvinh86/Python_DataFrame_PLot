```python
# EX2: Draw the graph of sin, cos, tan function within -2π -> 2π
```

```python
coordinates = np.linspace( - 2 * np.pi , 2 * np.pi , 20)
plt.plot(coordinates, np.sin(coordinates))
plt.plot(coordinates, np.cos(coordinates))
plt.plot(coordinates, np.tan(coordinates))
plt.legend(['sin', 'cos', 'tan'])
```

```python
# EX3: Draw the graph of  x = y ** 2 and x = y function on the same coordinate axis.
```

```python
coordinates = np.linspace(-10,10, 100)
plt.plot(coordinates, np.square(coordinates))
plt.plot(coordinates, coordinates)
plt.legend(['y=x**2', 'y=x'])
```

```python
# Sort solution:
df = pd.read_csv('/..../dientich.csv', 
                 sep=";", 
                 usecols=[0,5], 
                 header=None,
                 names=['Địa phương', "2015",],
                 skiprows=[0, 1, 2, 3], 
                 index_col=0)
df = df.dropna(0, how='all')
df = df.sort_values('2015')
topten = df.iloc[-16:-6, :]
topten.plot.bar()
```
